```python
def calc(a,b,c):
    add=a+b
    sub=a-b
    div=a/b
    mul=a*b
    if(c=='add'):
        return add
    elif(c=='sub'):
        return sub
    elif(c=='div'):
        return div
    elif(c=='mul'):
        return mul
```

```python
def leap(x):
    if(x%400==0) or (x%4==0 and x%100!=0):
        return str(x)+" is a leap year"
    else:
        return str(x)+" is not a leap year"
```

```python
leap(2000)
```
